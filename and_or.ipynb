# Logical AND (Conjunction) 

## Visualization

```python
fig, ax = plt.subplots()
ax.scatter([0, 0, 1], [0, 1, 0], color='blue')
ax.annotate('0', (0.02, 0))
ax.annotate('0', (0.02, 1))
ax.annotate('0', (1.02, 0))

ax.scatter([1], [1], color='red')
ax.annotate('1', (1.02, 1))

plt.plot([0.5, 1], [1, 0.5])

plt.title('AND Parameter Space with a Possible Linear Output Separation')
plt.savefig('imgs/and_space.png')
plt.show()
plt.close()
```

```python
X = np.transpose(np.array([[0, 0], [0, 1], [1, 0], [1, 1]]))
Y = np.array([[0, 0, 0, 1]])
```

```python
visualize_3D_err_space(X, Y, [2,1], nr_of_points_in_dir=20, act_function='relu')
```

```python
for seed in range(20):
    nn = MyNetwork([2,1], seed, weight_multiplier=2, weight_addition=-1)
    nn.fit(X, Y, num_iterations=20000, print_cost=False)
```

```python
misclassificationsAND = []
dim_list = [2,1]
weight_combs = init_weight_grid(dim_list, -1, 1, 20)
for i, weights in enumerate(weight_combs):
    print(f'Combination {i+1}/{len(weight_combs)}')
    nn = MyNetwork(dim_list, weights=weights)
    _, _, misclass = nn.fit(X, Y, learning_rate=0.1, num_iterations=2000, print_cost=False, plot_costs=False)
    misclassificationsAND.append(misclass)

print(f'Overall NR of Misclassification of 1000 models: {np.sum(misclassificationsAND)*4}')
```

# Logical OR (Disjunction) 

## Visualization

```python
fig, ax = plt.subplots()
ax.scatter([0], [0], color='blue')
ax.annotate('0', (0.02, 0))

ax.scatter([0, 1, 1], [1, 0, 1], color='red')
ax.annotate('1', (1.02, 0))
ax.annotate('1', (0.02, 1))
ax.annotate('1', (1.02, 1))

plt.plot([0, 0.5], [0.5, 0])

plt.title('OR Parameter Space with a Possible Linear Output Separation')
plt.savefig('imgs/or_space.png')
plt.show()
plt.close()
```

```python
X = np.transpose(np.array([[0, 0], [0, 1], [1, 0], [1, 1]]))
Y = np.array([[0, 1, 1, 1]])
```

```python
for seed in range(20):
    nn = MyNetwork([2,1], seed, weight_multiplier=2, weight_addition=-1)
    nn.fit(X, Y, num_iterations=20000, print_cost=False)
```
